# **Tuplas**

```python
#Retornando o primeiro elemento
tp[0]
```

```python
#Diferente das listas as tuplas são imutáveis, o que quewr dizer que não podemos alterar os seus elementos
tp[0] = "laranja"
```

```python
tp.count("maçã")
```

# **Dicionários**

```python
#para criar um dicionário utilizamos as {}
dc = {"maçã":20, "banana":10, "laranja":15, "uva":15} #Dicionários trabalham com o conceito de chave e valor
```

```python
dc
```

```python
#Acessando o valor de um dicionário através da chave
dc["maçã"]
```

```python
#Atualizando o valor da maçã
dc["maçã"] = 25
dc
```

```python
#Retornando todas as chaves do dicionário
dc.keys()
```

```python
#Retornando todas os valores do dicionário
dc.values()
```

```python
#Verificando se já existe uma chave no dicionário e caso não exista inserir
dc.setdefault("limão", 22)
```
